```python
pd.isnull(data5["age"]).values.ravel().sum()
```

#### age NaN será la media de la edad original
#### age NaN puede de igual forma ser reemplazado con el siguiente o anterior valido

```python
#Vamos a remplazar los 0 con la media de edad
data5["age"].fillna(df['age'].mean(),inplace=True);
data5["age"]
```
